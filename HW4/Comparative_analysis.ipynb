```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
out_paths = './ACGaN_Out_40/fretchet_distances.txt'
with open(out_paths, 'r') as file:
    ACGAN = [float(line.strip()) for line in file.readlines()]

out_paths = './DCGaN_Out_40/fretchet_distances.txt'
with open(out_paths, 'r') as file:
    DCGAN = [float(line.strip()) for line in file.readlines()]

out_paths = './WGaN_Out_40/fretchet_distances.txt'
with open(out_paths, 'r') as file:
    WGANGP = [float(line.strip()) for line in file.readlines()]
```

```python
plt.figure(figsize=(10,5))
plt.title("FID Scores comparsions for DCGAN vs WGAN vs ACGAN")
plt.plot(DCGAN,label="DCGAN",color='red')
plt.plot(WGANGP,label="WGAN",color='green')
plt.plot(ACGAN,label="ACGAN")
plt.xlabel("Epochs")
plt.ylabel("FID")
plt.legend()
plt.savefig('FID_comparison.jpg',
            format='jpeg',
            dpi=100,
            bbox_inches='tight')
plt.show()
```
